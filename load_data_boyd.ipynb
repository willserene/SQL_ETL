```python
import pandas as pd
from sqlalchemy import create_engine
from password import password
```

```python
game_penalties = "data/game_penalties.csv"
game_penalties_df = pd.read_csv(game_penalties)
game_penalties_df.head(5)
```

```python
game_skater_stats = "data/game_skater_stats.csv"
game_skater_stats_df = pd.read_csv(game_skater_stats)
game_skater_stats_df.head(5)
```

```python
engine = create_engine(f'postgresql://postgres:{password}@localhost:5432/hockey_db')

engine.table_names()
```

```python
game_penalties_df.to_sql(
    name='game_penalties',
    schema = 'public',
    con = engine,
    if_exists = 'append',
    index = False
)
```

```python
game_skater_stats_df.to_sql(
    name = "game_skater_stats",
    schema = 'public',
    con = engine,
    if_exists = 'append',
    index = False
    )
```
